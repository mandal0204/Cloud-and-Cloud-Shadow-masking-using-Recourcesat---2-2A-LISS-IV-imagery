```python
import rasterio
from rasterio.windows import Window
import os

def save_first_quarter(input_path, output_path):
    # Open the input TIFF file
    with rasterio.open(input_path) as src:
        width = src.width
        height = src.height

        # Define window for top-left quarter (you used one-third, so keeping that)
        window = Window(0, 0, width // 3, height // 3)

        # Read windowed data
        data = src.read(window=window)

        # Copy and update metadata
        profile = src.profile.copy()
        profile.update({
            "width": window.width,
            "height": window.height,
            "transform": src.window_transform(window)
        })

        # Ensure correct data type is preserved (important for mask files)
        profile['dtype'] = data.dtype

        # Write to new file
        with rasterio.open(output_path, 'w', **profile) as dst:
            dst.write(data)

# Example usage
save_first_quarter(
    r'/home/btech1/isro/dataset/train1/images/R2F27AUG2024069309009800054SSANSTUC00GTDA.tif',
    r'/home/btech1/isro/dataset/train1/R2F27AUG2024069309009800054SSANSTUC00GTDA.tif'
)

save_first_quarter(
    r'/home/btech1/isro/dataset/train1/images/RAF16SEP2024040353009000054SSANSTUC00GTDA.tif',
    r'/home/btech1/isro/dataset/train1/RAF16SEP2024040353009000054SSANSTUC00GTDA.tif'
)

save_first_quarter(
    r'/home/btech1/isro/dataset/train1/masks/mask_R2F27AUG2024069309009800054SSANSTUC00GTDA.tif',
    r'/home/btech1/isro/dataset/train1/mask_R2F27AUG2024069309009800054SSANSTUC00GTDA.tif'
)

save_first_quarter(
    r'/home/btech1/isro/dataset/train1/masks/mask_RAF16SEP2024040353009000054SSANSTUC00GTDA.tif',
    r'/home/btech1/isro/dataset/train1/mask_RAF16SEP2024040353009000054SSANSTUC00GTDA.tif'
)
```
